```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Starbucks_satisfactory_survey_encode_cleaned.csv')
data.head()
```

```python
#defining X
X = data[["age","income"]]

#Visualise data points
plt.scatter(X["age"],X["income"],c='black')
plt.xlabel('age')
plt.ylabel('income')
plt.show()
```

```python
K = 3 #number of clusters


# Select random observation as centroids
Centroids = (X.sample(n=K))
plt.scatter(X["age"],X["income"],c='black')
plt.scatter(Centroids["age"],Centroids["income"],c='red')
plt.xlabel('age')
plt.ylabel('income')
plt.show()
```
